### Task 1 — Data Exploration

First, we load the dataset and perform an initial inspection to understand its structure, including the shape, column types, and missing values.

We then conduct visual and statistical tests for normality and equal variance to validate the assumptions required for later hypothesis testing

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.power import TTestIndPower

df = pd.read_csv("top_100_movies_full_best_effort.csv")

# EDA
print("Dataset Shape:", df.shape)
print(df.info())
print("Missing Values:", df.isnull().sum())

df_clean = df.dropna(subset=['IMDb Rating', 'Box Office ($M)'])

# Assumption checks: Normality (Shapiro-Wilk) and Equal Variance (Levene)
# Checking IMDb Rating and Box Office distributions
shapiro_imdb = stats.shapiro(df_clean['IMDb Rating'])
shapiro_box = stats.shapiro(df_clean['Box Office ($M)'])

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df_clean['IMDb Rating'], kde=True)
plt.title('Distribution of IMDb Ratings')

plt.subplot(1, 2, 2)
sns.histplot(df_clean['Box Office ($M)'], kde=True)
plt.title('Distribution of Box Office Revenue')
plt.show()

print("IMDb Normality p-value: ", shapiro_imdb.pvalue)
print("Box Office Normality p-value: ", shapiro_box.pvalue)
```

## Task 2 -Hypothesis Testing

We formulate three research questions to explore relationships within the movie data. 
For each, we state the null (H0) and alternative (H1) hypotheses and justify the choice of test based on the distribution checks performed in Task 1.

### Research Question 1: Do movies with higher IMDb ratings earn more at the Box Office?

H0: No correlation between IMDb Rating and Box Office Revenue.

H1: Significant correlation between IMDb Rating and Box Office Revenue.

Justification: Using Spearman Correlation because Box Office data is not normally distributed.

```python
corr, p_val = stats.spearmanr(df_clean['IMDb Rating'], df_clean['Box Office ($M)'])
print("Spearman Correlation:", corr, "p-value:", p_val)
```

Conclusion:
?

### Research Question 2: Is there a difference in Box Office between 20th and 21st century movies?

H0: Mean Box Office is equal for both centuries.

H1: Mean Box Office differs between centuries.

```python
century_20 = df_clean[df_clean['Year'] < 2000]['Box Office ($M)']
century_21 = df_clean[df_clean['Year'] >= 2000]['Box Office ($M)']

#  Mann-Whitney U test used as data is non-normal.
u_stat, p_u = stats.mannwhitneyu(century_20, century_21)
print("Mann-Whitney U p-value:", p_u)
```

### Research Question 3: Is the count of Oscars won independent of the movie's language? (Categorical)

H0: Oscar wins and Language are independent.

H1: Oscar wins and Language are dependent.

```python
contingency = pd.crosstab(df['Language'], df['Oscars Won'] > 0)
chi2, p_chi, dof, ex = stats.chi2_contingency(contingency)
print("Chi-Square p-value:", p_chi)
```

## Task 3 - Confidence Intervals

We calculate the 95% confidence intervals for the mean IMDb Rating and mean Box Office revenue to estimate where the true population parameters likely reside. 

Visualizations are provided to illustrate these ranges clearly.

```python
imdb_mean = df_clean['IMDb Rating'].mean()
imdb_sem = stats.sem(df_clean['IMDb Rating'])
ci_imdb = stats.t.interval(0.95, len(df_clean['IMDb Rating'])-1, loc=imdb_mean, scale=imdb_sem)

box_mean = df_clean['Box Office ($M)'].mean()
box_sem = stats.sem(df_clean['Box Office ($M)'])
ci_box = stats.t.interval(0.95, len(df_clean['Box Office ($M)'])-1, loc=box_mean, scale=box_sem)

print("95% CI for IMDb Rating:", ci_imdb)
print("95% CI for Box Office ($M): ", ci_box)


plt.errorbar(['Mean IMDb Rating'], [imdb_mean], yerr=[[imdb_mean-ci_imdb[0]], [ci_imdb[1]-imdb_mean]], fmt='o', capsize=5)
plt.title('95% Confidence Interval for IMDb Rating')
plt.show()
```

## Task 4 — Power Analysis

A post-hoc power analysis is conducted for the comparison of Box Office revenue across centuries to evaluate if our sample size was sufficient to detect a meaningful effect at a standard 80% power level

```python
# Power Analysis for 2-sample T-test (as a proxy for evaluation)
analysis = TTestIndPower()
effect_size = 0.5  # Medium effect size
alpha = 0.05
power = 0.8

required_n = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=1.0)
print("Minimum sample size required per group:", required_n)

actual_power = analysis.power(effect_size=effect_size, nobs1=len(century_20), ratio=len(century_21)/len(century_20), alpha=alpha)
print("Actual Power achieved:", actual_power)
```
